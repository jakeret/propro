```python
def mem_hungry(size):
    a = []
    for i in range(size):
        a.append(np.random.random())
        
    b = []
    for i in range(size):
        t = []
        for j in range(size):
            t.append(i * a[j])
        b.append(t)

    b = np.array(b)
```
